```python
import pandas as pd

data = []
for i in range(0, 8):
    filename = f'./../../data/raw/migros_stores_data_{i}.csv'
    df = pd.read_csv(filename)
    del df['Unnamed: 0'] # unnecessary columns which needs to be deleted to ensure no duplications!?
    # we will need to process more (eg. empty city items!)
    data.append(df)
    
df = pd.concat(data, ignore_index=True)
df.drop_duplicates()
df.to_csv('./../../data/processed/migros_stores_data.csv')

print(df.info())
df.head(100)
```
